```python
import sys 
    
class BoxLetter:
    def __init__(self, destAddr, shippAddr, weight, sendMode):
        if sendMode not in ["normal", "express"]:
            print(sendMode+" n'est pas un mode d'envoi valide !")
            self.valid = False
        else:
            self.valid = True
            self.destAddr = destAddr
            self.shippAddr = shippAddr
            self.weight = weight
            self.sendMode = sendMode
    

    def toString(self):
        if self.valid:
            print("Adresse destination: {}".format(self.destAddr))
            print("Adresse expédition: {}".format(self.shippAddr))
            print("Poids: {} grammes".format(self.weight))
            print("Mode: {}".format(self.sendMode))

class Letter(BoxLetter):

    def __init__(self, destAddr, shippAddr, weight, sendMode, letterFormat):
        super().__init__(destAddr, shippAddr, weight, sendMode)
        if letterFormat not in ["A3", "A4"]:
            self.valid = False
            print(letterFormat+" n'est pas un format valide ! \n")
        else:
            self.letterFormat = letterFormat


    def calculTimbre(self):
        tarifDeBase = 0
        if self.letterFormat == "A4":
            tarifDeBase = 2.50
        elif self.letterFormat == "A3":
            tarifDeBase = 3.50

        if self.sendMode == "normal" or self.sendMode == "Normal":
            return tarifDeBase * (self.weight/1000)
        elif self.sendMode == "express" or self.sendMode == "Express":
            return tarifDeBase * (2.0 * (self.weight/1000))


    def toString(self):
        if self.valid:
            print("Lettre: ")
            super().toString()
            print("Format: {}".format(self.letterFormat))
            print("Prix du timbre: {} \n".format(self.calculTimbre()))



class Parcel(BoxLetter):

    def __init__(self, destAddr, shippAddr, weight, sendMode, volume):
        super().__init__(destAddr, shippAddr, weight, sendMode)
        self.volume = volume

    def calculTimbre(self):
        if self.sendMode == "normal" or self.sendMode == "Normal":
            return (0.25 * self.volume) * (self.weight/1000)
        elif self.sendMode == "express" or self.sendMode == "Express":
            return (0.25 * self.volume) * (self.weight/1000) * 2.0

    def toString(self):
        if self.valid:
            print("Colis: ")
            super().toString()
            print("Volume: {} litres".format(self.volume))
            print("Prix du timbre: {} \n".format(self.calculTimbre()))


L1 = Letter("Lille","Paris",80,"normal","A3")
L1.toString()

L2 = Letter("Marseille","Toulouse",80,"lente","A5")
L2.toString()

C1=Parcel("Marrakeche","Barcelone",3500,"express",2.25)
C1.toString()

C2=Parcel("Pékin","Madrid",3500,"lente",2.25)
C2.toString()
```
