```python
#Cration du filtre par saison 2019-20
df_2020 = df[df["season"]=="2019-20"]
df_2010 = df[df["season"]=="2009-10"]
df_2000 = df[df["season"]=="1999-00"]
df_1997 = df[df["season"]=="1997-98"]
```

```python
#Nombre de joueur par pays en 20
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(20,20))
sns.countplot(ax=axes[0,0], y="country", data=df_1997)
sns.countplot(ax=axes[0,1], y="country", data=df_2000)
sns.countplot(ax=axes[1,0], y="country", data=df_2010)
sns.countplot(ax=axes[1,1], y="country", data=df_2020)
```
