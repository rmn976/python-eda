```python
import pandas as pd 
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
print("- - - -lib loaded - - - ")
```

```python
df_hp = pd.read_csv('../datas/house_pricing.csv'); df_hp.head()
```

```python
df_hp.info()
```

```python
g = sns.relplot(x="SalePrice", y="LotArea", data=df_hp)
g.set_axis_labels("Prix de vente","Taille du lot"); 
plt.show();
```

```python
filter1 = df_hp['SalePrice'] < 500000
filter2 = df_hp['LotArea'] < 20000
df_hp2 = df_hp.where(filter1 & filter2)
```

```python
g2 = sns.relplot(x="SalePrice", y="LotArea", data=df_hp2)
g2.set_axis_labels("Prix de vente","Taille du lot"); 
plt.show();
```

```python
g3 = sns.relplot(x="LotFrontage", y="LotArea", data=df_hp)
g3.set_axis_labels("Taille maison","Taille du lot"); 
plt.show();
```

```python
filter3 = df_hp['LotFrontage'] < 200
filter4 = df_hp['LotArea'] < 100000
df_hp3 = df_hp.where(filter3 & filter4)
```

```python
g4 = sns.relplot(x="LotFrontage", y="LotArea", data=df_hp3)
g4.set_axis_labels("Taille maison","Taille du lot"); 
plt.show();
```
